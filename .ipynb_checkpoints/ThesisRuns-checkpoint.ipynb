## Centering the Galaxy

```python
# part = np.load("part.npy")
gas_position = part['gas']['position']
gas_velocity = part['gas']['velocity']
# host_position = [41792.15, 44131.23, 46267.68]
host_position = [42837.33, 41809.45, 39224.52]
centered_gas_position = gas_position-host_position
NHI = part['gas'].prop('number.density')
print(centered_gas_position)
print(NHI)
print(part.Cosmology)
```

## plotting the particles in 3D

```python

# Randomly select 100,000 particles
selected_indices = np.random.choice(centered_gas_position.shape[0], size=200000, replace=False)
selected_particles = centered_gas_position[selected_indices]
angular_momentum = np.cross(selected_particles,gas_velocity[selected_indices])
abs_angular_momentum = np.linalg.norm(angular_momentum, axis=1)
x = selected_particles[:,0]
y = selected_particles[:,1]
z = selected_particles[:,2]
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(x, y, z, marker='o',c=abs_angular_momentum,cmap='viridis',s=10, label='Particles')

# Enable interactive mode
# %matplotlib notebook
# %matplotlib widget
%matplotlib inline

# Customize the color mapping according to angular velocities
# cbar = plt.colorbar(scatter)
# cbar.set_label('Angular Velocity')

# You can customize the plot with labels, titles, etc.
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Particle Positions')
ax.view_init(elev=10, azim=90)  # Set the elevation and azimuthal angles as needed

# plt.legend()
plt.show()
```
